# Solution of question no. 1

1.	Scrape  	the  	details  of  	most  	viewed  videos  on  	YouTube  	from  	Wikipedia.  	Url 
 = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:  A) Rank   
B)	Name   
C)	Artist   
D)	Upload date   
E)	Views   


```python
# importing libraries
from selenium import webdriver
from time import sleep
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import requests
from selenium.common.exceptions import NoSuchElementException
```

```python
# choosing Chrome as our browser
driver=webdriver.Chrome()
```

```python
# getting in to web page
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python

driver.maximize_window()
```

```python
# variables for storing scraped data
Rank=[]
Name=[]
Artist=[]
Upload_date=[]
Views=[]
```

```python
# scraping all atributes as per question
    
try:
    name=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[1]')
    for i in name:
        Name.append(i.text)
        
except NoSuchElementException:
    Name.append("-")
    
    
try:
    artist=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[2]')
    for k in artist:
        Artist.append(k.text)
        
except NoSuchElementException:
    Artist.append("-")
    
    
try:
    upload_date=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[4]')
    for l in upload_date:
        Upload_date.append(l.text)
        
except NoSuchElementException:
    Upload_date.append("-")
    

try:
    views=driver.find_elements(By.XPATH,'//table[@class="sortable wikitable sticky-header static-row-numbers sort-under col3center col4right jquery-tablesorter"]/tbody/tr/td[3]')
    for m in views:
        Views.append(m.text)
        
except NoSuchElementException:
    Views.append("-")
```

```python
#checking lengths
print(len(Name),len(Artist),len(Upload_date),len(Views))
```

```python
#creating the dataframe
df1=pd.DataFrame({"Name":Name,"Artist":Artist,"Upload_date":Upload_date,"Views":Views})
df1
```

B)	Author name   
C)	Volumes sold   
D)	Publisher   
E)	Genre   
 
Url - https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare  


```python
#importing libraries
import pandas as pd
from selenium import webdriver
from time import sleep
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import requests
from selenium.common.exceptions import NoSuchElementException, ElementNotInteractableException
```

```python
#opening chrome browser
driver=webdriver.Chrome()
```

```python
# getting in to webpage
driver.get("https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare")
```

```python
driver.maximize_window()
```

```python
#for storing scraped data
Book_name = []
Author_name = []
Volumes_sold = []
Publisher = []
Genre = []
```

```python
#finding and storing our data

try:
    book=driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[2]')
    for i in book:
        Book_name.append(i.text)
        
except NoSuchElementException:
    Book_name.append("-")
    sleep(1)
    
    
try:
    author_name=driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[3]')
    for j in author_name:
        Author_name.append(j.text)
        
except NoSuchElementException:
    Author_name.append("-")
    sleep(1)
    
try:
    volumes_sold=driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[4]')
    for k in volumes_sold:
        Volumes_sold.append(k.text)
        
except NoSuchElementException:
    Volumes_sold.append("-")
    sleep(1)
    
try:
    publisher=driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[5]')
    for l in publisher:
        Publisher.append(l.text)
        
except NoSuchElementException:
    Publisher.append("-")
    sleep(1)
    
try:
    genre=driver.find_elements(By.XPATH,'//table[@class="in-article sortable"]/tbody/tr/td[6]')
    for m in genre:
        Genre.append(m.text)
        
except NoSuchElementException:
    Genre.append("-")
    sleep(1)    
```

```python
print(len(Book_name),len(Author_name),len(Volumes_sold),len(Publisher),len(Genre))
```

```python
# creating DataFrame
df6= pd.DataFrame({"Book_name":Book_name,"Author_name":Author_name,"Volumes_sold":Volumes_sold,
                  "Publisher":Publisher,"Genre":Genre})
df6
```

8.	Details of Datasets from UCI machine learning repositories.   
Url = https://archive.ics.uci.edu/  You have to find the following details:   
A)	Dataset name   
B)	Data type   
C)	Task   
D)	Attribute type   
E)	No of instances   
F)	No of attribute 
G) Year   
 
Note: - from the home page you have to go to the Show All Dataset page through code.   


```python
#importing libraries
import pandas as pd
from selenium import webdriver
from time import sleep
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import requests
from selenium.common.exceptions import NoSuchElementException
```

```python
#opening chrome browser
driver=webdriver.Chrome()
```

```python
driver.maximize_window()
```

```python
# getting in to the webpage
driver.get("https://archive.ics.uci.edu/")
```

```python
all_dataset=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div/div/div/a[1]")
all_dataset.click()
```

```python
# collecting urls for scraping indivisualy each 
urls=[]

url=driver.find_elements(By.XPATH,'//a[@class="link-hover link text-xl font-semibold"]')
for i in url:
    urls.append(i.get_attribute("href"))
    sleep(1)
    
```

```python
len(urls)
```

```python
#taking variables for storing scraped data
Dataset_name = []   
Data_type  = [] 
Task  = [] 
Attribute_type = []  
No_of_instances = []  
No_of_attribute = []
Year = []
```

```python

for a in  urls:         # taking scraped urls one by one
    driver.get(a)
    sleep(5)

# extracting data from webpage
    try:
        dataset_name=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[1]/div[2]/div/h1")
        Dataset_name.append(dataset_name.text)
        sleep(1)
            
    except NoSuchElementException:
        Dataset_name.append("-")
        
        
    try:
        data_type=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[1]/p")
        Data_type.append(data_type.text)
            
    except NoSuchElementException:
        Data_type.append("-")
    
    try:
        task=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[3]/p")
        Task.append(task.text)
            
    except NoSuchElementException:
        Task.append("-")
    
    
    try:
        attribute_type=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[4]/p")
        Attribute_type.append(attribute_type.text)
            
    except NoSuchElementException:
        Attribute_type.append("-")
    
   
    try:
        no_of_instances=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[5]/p")
        No_of_instances.append(no_of_instances.text)
            
    except NoSuchElementException:
        No_of_instances.append("-")
        
        
    try:
        no_of_attribute=driver.find_element(By.XPATH,"/html/body/div/div[1]/div[1]/main/div/div[1]/div[1]/div[2]/div[2]/div[6]/h1")
        No_of_attribute.append(no_of_attribute.text)
        sleep(1)
            
    except NoSuchElementException:
        No_of_attribute.append("-")
    
    
    try:
        year=driver.find_element(By.XPATH,'//h2[@class="text-sm text-primary-content"]')
        Year.append(year.text)
        sleep(1)
            
    except NoSuchElementException :
        Year.append("Not_present")
        
        
        
```

```python
# checking length of all attributes
print(len(Dataset_name),len(Data_type),len(Task),len(Attribute_type),len(No_of_instances),len(No_of_attribute),len(Year))
```

```python
#creating DataFrame
df8=pd.DataFrame({"Dataset_name":Dataset_name,"Data_type":Data_type,"Task":Task,"Attribute_type":Attribute_type,
                 "No_of_instances":No_of_instances,"No_of_attribute":No_of_attribute,"Year":Year})
df8
```
